## what is the  number of orders per state?

## what is the total reveny per city?

```python
rev_city=df.groupby('city')['Price Each'].sum().sort_values(ascending=False).reset_index()
rev_city
```

```python
px.bar(rev_city,'city',y='Price Each',labels={'Price Each' :'Total Rev'},text_auto=True)
```
